```python
# Step 2: Import necessary modules and open a webpage

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Create a Chrome browser instance
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Maximize the window
driver.maximize_window()

# Open a simple website 
driver.get("https://practicetestautomation.com/practice-test-login/")

# Wait for a few seconds so you can see it open
time.sleep(3)

# Close the browser
driver.quit()
```

```python
# Step 3: Automated login test with Selenium

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Create browser instance
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.maximize_window()

# Step 1: Open the login page
driver.get("https://practicetestautomation.com/practice-test-login/")
time.sleep(2)

# Step 2: Enter username
username_box = driver.find_element(By.ID, "username")
username_box.send_keys("student")

# Step 3: Enter password
password_box = driver.find_element(By.ID, "password")
password_box.send_keys("Password123")

# Step 4: Click the submit button
login_button = driver.find_element(By.ID, "submit")
login_button.click()

# Step 5: Wait and check result
time.sleep(2)
success_message = driver.find_element(By.XPATH, "//h1[contains(text(),'Logged In Successfully')]")
print("✅ Test Passed! Message found:", success_message.text)

# Step 6: Close browser
time.sleep(2)
driver.quit()
```
